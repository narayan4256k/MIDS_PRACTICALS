```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the 8 points from your problem statement
P = np.array([
    [0.1, 0.6],  # P1
    [0.15, 0.71], # P2
    [0.08, 0.9],  # P3
    [0.16, 0.85], # P4
    [0.2, 0.3],   # P5
    [0.25, 0.5],  # P6
    [0.24, 0.1],  # P7
    [0.3, 0.2]    # P8
])

# 2. Set initial centroids as m1=P1 and m2=P8
m1 = P[0]
m2 = P[7]

print(f"Initial Centroids: m1={m1}, m2={m2}\n")

# 3. K-Means Algorithm
for i in range(100):
    C1, C2 = [], []
    
    # Assignment Step
    for point in P:
        dist1 = np.linalg.norm(point - m1)
        dist2 = np.linalg.norm(point - m2)
        if dist1 < dist2:
            C1.append(point)
        else:
            C2.append(point)
    
    # Update Step (Calculate new means)
    m1_new = np.mean(C1, axis=0)
    m2_new = np.mean(C2, axis=0)
    
    # Check for convergence
    if np.allclose(m1, m1_new) and np.allclose(m2, m2_new):
        break
    
    m1, m2 = m1_new, m2_new

# 4. Final Outputs for your Journal
print("--- Final Results ---")
print(f"Cluster 1: {C1}")
print(f"Population of Cluster 1: {len(C1)}")
print(f"Cluster 2: {C2}")
print(f"Population of Cluster 2: {len(C2)}")
print(f"\nUpdated m1: {m1}")
print(f"Updated m2: {m2}")

# 5. Graphical Representation
C1 = np.array(C1)
C2 = np.array(C2)
plt.scatter(C1[:,0], C1[:,1], color='blue', label='Cluster 1')
plt.scatter(C2[:,0], C2[:,1], color='red', label='Cluster 2')
plt.scatter([m1[0], m2[0]], [m1[1], m2[1]], color='black', marker='X', s=100, label='Centroids')
plt.legend()
plt.title("K-Means Clustering Result")
plt.show()
```
