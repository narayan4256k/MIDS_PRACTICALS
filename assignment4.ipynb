```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sm

try:
    # 1. Load the dataset
    abalone = pd.read_csv('abalone.csv')
    print("✅ Dataset loaded successfully!")
    
    # 2. Encode the 'Sex' column (Converts 'M', 'F', 'I' into numbers 0, 1, 2)
    le = LabelEncoder()
    abalone['Sex'] = le.fit_transform(abalone['Sex'])
    
    # --- PART A: CLASSIFICATION (Predicting Rings) ---
    print("\n--- Part A: Classification (Predicting Rings) ---")
    y = abalone['Rings'].values
    x = abalone.drop('Rings', axis=1).values
    
    # Split the data
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
    
    # Train Naive Bayes Classifier
    classifier = GaussianNB()
    classifier.fit(x_train, y_train)
    y_pred = classifier.predict(x_test)
    
    print(f"Accuracy of classification: {accuracy_score(y_test, y_pred) * 100:.2f}%")
    
    # --- PART B: LINEAR REGRESSION (Predicting Age) ---
    print("\n--- Part B: Linear Regression (Predicting Age) ---")
    # Create the new 'age' attribute (Rings + 1.5)
    abalone['age'] = abalone['Rings'] + 1.5
    
    y_age = abalone['age'].values
    # We drop both Rings and age from the input features to prevent cheating
    x_age = abalone.drop(['Rings', 'age'], axis=1).values 
    
    x_train_age, x_test_age, y_train_age, y_test_age = train_test_split(x_age, y_age, test_size=0.25, random_state=0)
    
    # Train Linear Regression Model
    regressor = LinearRegression()
    regressor.fit(x_train_age, y_train_age)
    y_pred_age = regressor.predict(x_test_age)
    
    print(f"Train Score: {regressor.score(x_train_age, y_train_age):.4f}")
    print(f"Test Score: {regressor.score(x_test_age, y_test_age):.4f}")
    print(f"Mean absolute error = {round(sm.mean_absolute_error(y_test_age, y_pred_age), 2)}")
    print(f"Mean squared error = {round(sm.mean_squared_error(y_test_age, y_pred_age), 2)}")
    print(f"R2 score = {round(sm.r2_score(y_test_age, y_pred_age), 2)}")

except FileNotFoundError:
    print("❌ ERROR: Could not find 'abalone.csv'. Make sure the file is in the same folder as this notebook!")
except KeyError as e:
    print(f"❌ ERROR: Column {e} not found. Ensure your CSV has headers like 'Sex', 'Length', 'Diameter', etc.")
```
