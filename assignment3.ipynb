```python
import nltk
import random

# Download the necessary dataset from NLTK
nltk.download('movie_reviews')
from nltk.corpus import movie_reviews

print("Loading and shuffling document data...")
# Load the dataset
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]

random.shuffle(documents)

# Extract features (get the most common 2000 words)
print("Extracting word features...")
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())
word_features = list(all_words)[:2000]

def document_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words)
    return features

# Prepare the training and testing sets
print("Training the Naive Bayes Classifier...")
featuresets = [(document_features(d), c) for (d, c) in documents]
train_set, test_set = featuresets[100:], featuresets[:100]

# Train the Classifier
classifier = nltk.NaiveBayesClassifier.train(train_set)

# Test the classifier and print accuracy
accuracy = nltk.classify.accuracy(classifier, test_set) * 100
print(f'\n✅ Accuracy of Naive Bayes Classifier is {accuracy:.2f}%\n')

# List the most informative features
print("Top 5 Most Informative Features:")
classifier.show_most_informative_features(5)
```
