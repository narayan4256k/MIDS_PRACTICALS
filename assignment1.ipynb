```python
import pandas as pd

try:
    # 1. Importing the dataset
    df = pd.read_csv("train.csv")
    
    print("--- 1. Initial Dataset Info ---")
    df.info()
    
    # 2. Filling the missing values in 'Age' with the mean of the column
    df['Age'] = df['Age'].fillna(df['Age'].mean())
    print("\n✅ Filled missing 'Age' values with the mean.")
    
    # 3. Dropping the 'Cabin' column because it has too many missing values
    df = df.drop(['Cabin'], axis=1)
    print("✅ Dropped the 'Cabin' column.")
    
    print("\n--- 2. Cleaned Dataset Info ---")
    df.info()
    
except FileNotFoundError:
    print("❌ ERROR: Could not find 'train.csv'. Make sure the file is in the same folder as this notebook!")
```
